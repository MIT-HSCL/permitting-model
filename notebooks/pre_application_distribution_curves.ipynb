# Pre-application duration distributions

Compares the **PDF/CDF** of pre-application timing options used in the simulation (`baseline`, `lognormal_180`, `lognormal_60`, `poisson_10`, etc.).

No simulation run — this is a distribution visualization notebook.


## 1. Setup


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

LOGNORMAL_SIGMA = 0.896

# Use fixed LA lognormal parameters provided by user.
shape_la = 0.8961865652479636
loc_la = 0.0
scale_la = 455.8128768371504

print("LA fitted lognormal parameters:")
print(f"  shape (sigma): {shape_la:.4f}")
print(f"  loc: {loc_la:.4f}")
print(f"  scale: {scale_la:.4f}")
print(f"  implied median: {scale_la:.2f} days")
```

## 2. Compare PDF curves


```python
# X range for comparing curves
x = np.linspace(0, 1500, 2000)

# Scenario distributions used in the simulation settings
pdf_180 = lognorm.pdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=180)
pdf_60 = lognorm.pdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=60)
pdf_10 = lognorm.pdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=10)

# Fitted LA distribution
pdf_la_fit = lognorm.pdf(x, s=shape_la, loc=loc_la, scale=scale_la)

plt.figure(figsize=(11, 6))
plt.plot(x, pdf_180, label="lognormal_180 (sigma=0.53)", linewidth=2)
plt.plot(x, pdf_60, label="lognormal_60 (sigma=0.53)", linewidth=2)
plt.plot(x, pdf_10, label="lognormal_10 (sigma=0.53)", linewidth=2)
plt.plot(x, pdf_la_fit, label="LA fitted lognormal", linewidth=3, linestyle="--", color="black")

plt.xlabel("Days")
plt.ylabel("Density")
plt.title("Pre-Application Distribution Comparison (PDF)")
plt.xlim(0, 500)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. Compare CDF curves (optional)


```python
# Optional: CDF comparison (useful for interpreting timing milestones)
cdf_180 = lognorm.cdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=180)
cdf_60 = lognorm.cdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=60)
cdf_10 = lognorm.cdf(x, s=LOGNORMAL_SIGMA, loc=0, scale=10)
cdf_la_fit = lognorm.cdf(x, s=shape_la, loc=loc_la, scale=scale_la)

plt.figure(figsize=(11, 6))
plt.plot(x, cdf_180, label="lognormal_180 (sigma=0.896)", linewidth=2)
plt.plot(x, cdf_60, label="lognormal_60 (sigma=0.896)", linewidth=2)
plt.plot(x, cdf_10, label="lognormal_10 (sigma=0.896)", linewidth=2)
plt.plot(x, cdf_la_fit, label="LA fitted lognormal", linewidth=3, linestyle="--", color="black")

plt.xlabel("Days")
plt.ylabel("Cumulative probability")
plt.title("Pre-Application Distribution Comparison (CDF)")
plt.xlim(0, 1500)
plt.ylim(0, 1.02)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
